```python
# the module below will allow us to perform linear regression
import statsmodels.api as sm
```

The function `lin_reg(x,y)` given below performs ordinary least squares (OLS) linear regression using `sm.OLS` from the `statsmodels.api` module.

The code enclosed in `''' '''` is the docstring of the function `lin_reg`.

`x` in the `lin_reg` function is a matrix that contains the regressors, and `y` represents the vector containing the dependent variable. Note that `x` might contain one vector or multiple vectors. In the case that `x` contains one vector $x$, the regression gives:

$$ y = \beta_0 + \beta_1 x $$

In the case that `x` contains multiple vectors $x_1, \dots, x_k$, the regression becomes:

$$ y = \beta_0 + \beta_1 x_1 + \dots + \beta_k x_k$$

The $\beta$'s are the regression coefficients obtained using least squares. Note that `sm.add_constant` is used in the function below to make `x` look like the matrix $A$ we use in least squares, whose first column contains all ones.

```python
def lin_reg(x, y):
    '''
    ordinary linear regression using least-squares
    
    Parameters
    ----------  
    x: regressors (numpy array)
    y: dependent variable (numpy array)
    
    Returns
    -------
    coefficients: regression coefficients (pandas Series)
    residuals: regression residuals (numpy array)
    r_squared: correlation coefficient (float)
    
    '''
    x = sm.add_constant(x)
    model = sm.OLS(y, x).fit()
    coefficients = model.params
    residuals = model.resid
    r_squared = model.rsquared
    return coefficients, residuals, r_squared
```

### Let `y` be the Apple stock returns, and `x` be the Intel stock returns. Use the `lin_reg` function defined above to find $y=\beta_0 + \beta_1 x$. 

```python
# code for question 10

# reshaping to 2d so can use hstack later 
aapl = returns['AAPL'].to_numpy().reshape(-1,1)
intc = returns['INTC'].to_numpy().reshape(-1,1)
msft = returns['MSFT'].to_numpy().reshape(-1,1)
ibm = returns['IBM'].to_numpy().reshape(-1,1)

betas, res, R2 = lin_reg(intc,aapl)
print('The linear regression equation is y={:.5f}+{:.3f}x'.format(betas[0],betas[1]))
```

### 11. Plot the cumulative sum of the Apple returns prediction from least squares on top of the cumulative sum of the actual Apple returns. How well do the Intel stock returns describe the Apple stock returns? Answer this question using a quantitative measure of choice to describe how well the regression describes the actual data (you should research what is standard procedure in answering these types of questions).

```python
# code for question 11

aapl_cumsum = aapl.cumsum()
aapl_preds_daily = betas[0]+betas[1]*intc
aapl_preds_cumsum = aapl_preds_daily.cumsum()

SSR = ((aapl_cumsum-aapl_preds_cumsum)**2).sum()
SST = ((aapl_cumsum-aapl_cumsum.mean())**2).sum()
R2 = 1-SSR/SST
print('R squared for the regression fit: {:.3f}'.format(R2))
```

```python
plt.figure(figsize=(9,6))
plt.plot(aapl_cumsum,label='Actual',lw=2)
plt.plot(aapl_preds_cumsum,label='Predicted',lw=2)
plt.title('Apple actual vs predicted cumulative return',fontsize=fs)
plt.ylabel('Cumulative return',fontsize=fs-2)
plt.xlabel('Day',fontsize=fs-2)
plt.legend(prop={'size': 12})
plt.show()
```

**ANSWER FOR QUESTION 11**: (double click here and type your answer)

The statistical measure that I used was the coefficient of determination or $R^2$ which gives an indication of how good the predictions from the regression model estimate the actual data and can be interpreted as the variability of the target that is explained by the model. Generally, $R^2$ is within the range of 0 and 1 with a $R^2=1$ meaning that the regression fit models the data perfectly. Here, the regression model obtained an $R^2 = 0.934$, i.e., $93.4\%$ of the variability in the Apple cumulative return was accounted for by the regression model. As this value is very close to 1, the regression model using the Intel stock returns as the explanatory variable describes the Apple cumulative return pretty well.

**ANSWER FOR QUESTION 24**: (double click here and type your answer)

Let $X$ be our $252 \times 488$ matrix of data. Then it must have SVD of $X = U\Sigma V^{\textrm{T}}$.

Also, $X$ has (pop) covariance $C(X) =\frac{X^{\textrm{T}}X}{252}.$ However, if we standardize the columns of $X$ (as we did above), then $C(X) = \frac{X^{\textrm{T}}X}{252}$ would actually give us the correlation matrix. Thus from here on, we will refer to $C(X)$ as the correlation matrix. 


Then by doing matrix multiplication we have the following
(noting that since the columns of $U$ are orthogonal, $ U^{\textrm{T}} U = I$)

$C(X) = \frac{X^{\textrm{T}}X}{252} \\
= (U\Sigma V^{\textrm{T}})^{\textrm{T}}U\Sigma V^{\textrm{T}}/252 \\
= V\Sigma U^{\textrm{T}} U\Sigma V^{\textrm{T}}/252 \\
= V \Sigma^2 V^{\textrm{T}}/252
$

which means that the singular values $\sigma_i$ of the standardized $X$ are related to the eigenvalues of the correlation $C(X)$ via the following equation:

$\lambda_i = \frac{\sigma_i^2}{252}.$

(I checked this result by showing above this is true for the first singular vaue and eigenvalue.)

### 26. How are the eigenvectors of the correlation matrix `eigvecs_corr` connected to the left singular vectors of the standardized returns matrix in Question 22? Here you are expected to derive the relation theoretically.

**ANSWER FOR QUESTION 26**: (double click here and type your answer)

Similar to the matrix multiplication in question 24, we can show the following for $U$ using the SVD of the standardized $X$ 


$XX^{\textrm{T}}U = U\Sigma V^{\textrm{T}}(U\Sigma V^{\textrm{T}})^{\textrm{T}} U \\
=U\Sigma V^{\textrm{T}}V\Sigma U^{\textrm{T}}U \\
=U\Sigma^2
$

Now since $U$ is a diagonal matrix, we have

$XX^{\textrm{T}}U =\Sigma^2 U$ 

and considering each column of $U$, $u_i$, by itself would give us 

$XX^{\textrm{T}}u_i =\sigma_i^2 u_i$ 

(a set of eigenvalue equations) which means that the left singular vectors of the standardized returns matrix are the eigenvectors of $XX^{\textrm{T}}$.
